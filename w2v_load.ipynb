## Load Model

```python
from gensim.models import Word2Vec
import pickle

filename = 'w2v.model'
file = open(filename, 'rb')
model = pickle.load(file)
file.close()
```

## Activation-like words

```python
model.wv.most_similar(positive=['activate', 'activation', 'activator', 'stimulatory', 'stimulation', 'stimulate', 'increase'], topn=200)
```

## Inhibit-like words

```python
model.wv.most_similar(positive=['inhibit', 'inhibitors', 'inhibitor', 'inhibition', 'repress', 'repression', 'repressor', 'block'], topn=100)
```

## Both

```python
model.wv.most_similar(positive=['activate', 'activation', 'activator', 'stimulatory', 'stimulation', 'stimulate', 'increase', 'inhibit', 'inhibitors', 'inhibitor', 'inhibition', 'repress', 'repression', 'repressor', 'block'], topn=200)
```

## One-sided Positive

```python
model.wv.most_similar(positive=['activate', 'activation', 'activator', 'stimulatory', 'stimulation', 'stimulate', 'increase'], negative=['inhibit', 'inhibitors', 'inhibitor', 'inhibition', 'repress', 'repression', 'repressor', 'block'], topn=50)
```

## One-sided Negative

```python
model.wv.most_similar(negative=['activate', 'activation', 'activator', 'stimulatory', 'stimulation', 'stimulate', 'increase'], positive=['inhibit', 'inhibitors', 'inhibitor', 'inhibition', 'repress', 'repression', 'repressor', 'block'], topn=50)
```

## apc/c

```python
model.wv.most_similar('apc/c', topn=50)
```
